# Selenium




So far we have avoided Selenium with the following techniques:

1. Find source 
2. Build URLs
3. Select options in pulldowns
4. Find the JSON response

Those steps work for 95 percent of sites. For the 5 percent:

5. When all else falls short...Selenium as a last resort.

### What is Selenium?

#### An open source automated framework to test and validate applications/websites.

We interface with it using Python.

Several elements to the framework:
- Selenium Integrated Development Environment
- Selenium Grid
- Webdriver (provides ablity to interact with modern browsers)
